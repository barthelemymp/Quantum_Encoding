```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
```

```python
train = {}
valid = {}
acc = {}
legendname = {}
for w in range(2,6):
    for b in range(1,4):
        file_name = 'Iris_w'+str(w)+'_b'+str(b)+'_model_id'
        legendname[file_name] = str(w) +' qubits and '+str(b)+' classifying layers'
        train[file_name] = np.load(file_name+'train_losses.npy')
        valid[file_name] = np.load(file_name+'valid_losses.npy')
        acc[file_name] = np.load(file_name+'valid_acc.npy')

for b in range(1,4):
    file_name = 'IrisAEQC_w'+str(2)+'_b'+str(b)+'_model_id'
    legendname[file_name] = 'Amplitude encoding and '+str(b)+' classifying layers'
    train[file_name] = np.load(file_name+'train_losses.npy')[:10]
    valid[file_name] = np.load(file_name+'valid_losses.npy')[:10]
    acc[file_name] = np.load(file_name+'valid_acc.npy')[:10]
```

```python
x = list(range(1,10))
for w in range(2,6):
    for b in range(1,4):
        file_name = 'Iris_w'+str(w)+'_b'+str(b)+'_model_id'
        
        plt.plot(x,train[file_name][1:] , linewidth=3, marker='o', label=legendname[file_name])
        
        
        
        
#         # The plot with line styling
# plt.plot(x_points, y1_points, color='tab:blue', linestyle='-', linewidth=3, marker='o', label="cosine")
# plt.plot(x_points, y2_points, color='tab:red', linestyle='--', linewidth=2, marker='x', label="sine")
legend_x = 1.05
legend_y = 1.15
# Add title, labels, etc.
plt.title("Convergence of training loss", fontsize=22)
plt.xlabel("epochs", fontsize=18)
plt.ylabel("Loss on training set", fontsize=18)
plt.legend(fontsize=16, bbox_to_anchor=(legend_x, legend_y))
```
